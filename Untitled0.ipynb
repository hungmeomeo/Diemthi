```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import pickle
import warnings
import os
import statistics as st
from math import nan, isnan
from matplotlib.offsetbox import AnchoredText
```

```python
path  = f"/content/drive/My Drive/Colab Notebooks/file csv"
df  = pd.read_csv(path+'/diemthi.csv')
```

```python

df = df.replace('None', np.nan, regex=True)
df = df.astype(float)


toan = df['Toan'].to_list()
ly = df['Ly'].to_list()
hoa = df['Hoa'].to_list()
sinh = df['Sinh'].to_list()
anh = df['Anh'].to_list()
van = df['Van'].to_list()


df['A1'] = df[['Toan', 'Ly', 'Anh']].sum(axis = 1, skipna=False)
A1 = df['A1'].to_list()
df['A0'] = df[['Toan', 'Ly', 'Hoa']].sum(axis = 1, skipna=False)
A0 = df['A0'].to_list()
df['B'] = df[['Toan', 'Hoa', 'Sinh']].sum(axis = 1, skipna=False)
B = df['B'].to_list()
df['D7'] = df[['Toan', 'Van', 'Anh']].sum(axis = 1, skipna=False)
D7 = df['D7'].to_list()


print(df)


```

```python
num1 = np.arange(0, 31, 1,dtype =float)

def plot(khoi, s):
  fig, ax = plt.subplots(1, 1, figsize=(20, 15))

  arr = ax.hist(khoi, ec ='white', bins =num1)
  plt.xticks(num1)
  # Set title
  ax.set_title("Title")
    
  # adding labels
  ax.set_xlabel('Điểm thi')
    
  # Make some labels.
  rects = ax.patches
  labels = [arr[0][i] for i in range(len(rects))]
    
  for rect, label in zip(rects, labels):
      height = rect.get_height()
      ax.text(rect.get_x() + rect.get_width() / 2, height+0.01, int(label),
              ha='center', va='bottom')
  anchor_text = ("Điểm cao nhất: "+str(round(np.nanmax(khoi),2))+"\n"+
  "Điểm thấp nhất: "+str(np.nanmin(khoi))+"\n"+
  "Điểm trung bình: "+ str(round(st.mean([x for x in khoi if isnan(x) == False]), 2))+"\n"+
  "Điểm trung vị: "+ str(round(st.median([x for x in khoi if isnan(x) == False]),2)))

  # Show plot
  plt.title("Phổ điểm khối "+s)
  at = AnchoredText(
    anchor_text, prop=dict(size=15), frameon=True, loc='upper left')
  at.patch.set_boxstyle("round,pad=0.,rounding_size=0.2")
  ax.add_artist(at)
  plt.savefig("/content/drive/My Drive/Colab Notebooks/image/diemthi"+s+".png")
  plt.savefig(s+".png")
  plt.show()
```

```python
plot(A1, "A1")
plot(A0, "A0")
plot(B, "B")
plot(D7, "D1")
```
